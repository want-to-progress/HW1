```
# change image size
size = (128, 128)
```

#Load CSV

```
labels = pd.read_csv("training_labels.csv")
print(labels.columns)
# add labels to dict
# labels to index
label_dic = dict()
cnt = 0
for i in range(labels.shape[0]):
    if (label_dic.get(labels['label'][i]) is None):
        label_dic[labels['label'][i]] = cnt
        cnt += 1
print(label_dic)
```

#Training Data

```
input_arr = np.array([])
label_arr = np.array([])
# load image
for i in range(int(labels.shape[0])):
    image = img.load_img("training_data/"+str(labels["id"][i]).zfill(6)+'.jpg',
                         target_size=size)
    image = img.img_to_array(image)/255
    # save as numpy array
    if (i == 0):
        input_arr = np.array([image])
        label_arr = np.array([label_dic[labels["label"][i]]])
    else:
        input_arr = np.concatenate((input_arr, np.array([image])))
        label_arr = np.concatenate((label_arr,
                                    np.array([label_dic[labels["label"][i]]])))
print(input_arr.shape)
print(label_arr.shape)
```

```
np.save('train_data', input_arr)
np.save('train_labels', label_arr)
```

#Testing Data

```
filenames = os.listdir('testing_data')
input_arr = np.array([])
id_arr = np.array([])
# load image
for name in filenames:
    image = img.load_img("testing_data/"+name, target_size=size)
    image = img.img_to_array(image)/255
    # save as numpy array
    if (input_arr.shape[0] == 0):
        input_arr = np.array([image])
        id_arr = np.array([int(name.split('.')[0])])
    else:
        input_arr = np.concatenate((input_arr, np.array([image])))
        id_arr = np.concatenate((id_arr, np.array([int(name.split('.')[0])])))
print(input_arr.shape)
```

```
np.save('test', input_arr)
np.save('test_ids', id_arr)
```
